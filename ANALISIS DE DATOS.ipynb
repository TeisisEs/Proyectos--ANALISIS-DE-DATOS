## ¿Que es la esperanza de vida de un pais?

La esperanza de vida es un indicador demográfico que estima el promedio de años que una persona podría vivir, partiendo de las condiciones actuales de mortalidad en una población específica. Este indicador se calcula considerando factores como la tasa de mortalidad en diferentes grupos de edad y refleja las condiciones generales de salud, desarrollo socioeconómico y acceso a servicios básicos, como atención médica y saneamiento. Es una medida clave para evaluar la calidad de vida y el bienestar en una sociedad, así como para comparar el progreso entre distintos países o regiones.


## Esperanza de vida en El Salvador panorama General

En 2022 la esperanza de vida en El Salvador subió hasta llegar a 71,48 años.

Ese año la esperanza de vida de las mujeres fue de 75,81 años, mayor que la de los hombres que fue de 66,84 años.

El Salvador ha empeorado su situación respecto a los 194 países de los que publicamos la Esperanza de vida y ha pasado de ocupar el puesto 107 en 2021 al 109 en 2022. Esto quiere decir que es un país con una baja esperanza de vida respecto al resto de los países del mundo.

Si miramos la evolución de la Esperanza de Vida en El Salvador en los últimos años, vemos que ha subido respecto a 2021 en el que fue de 70,75 años, al contrario de lo que ocurre respecto a 2012, en el que estaba en 71,83 años.

## PREPARACION DE LOS DATOS

### Esperanza de Vida

```python
import pandas as pd

# Cargar el archivo CSV 
df = pd.read_csv('esperanza_vida_el_salvador.csv')

# Modificar los encabezados de las columnas
df.columns = ['AÑOS', 'MUJERES', 'HOMBRES','ESPERANZA DE VIDA']

# Formatear columnas específicas con un separador de miles
columnas_a_formatear = ['MUJERES', 'HOMBRES', 'ESPERANZA DE VIDA']

for columna in columnas_a_formatear:
    df[columna] = df[columna].map(lambda x: f"{x / 100:,.2f}".replace(".", ","))


# Revisar los datos
df.head(11)
```

### Homicidios ANTES del Regimen

```python
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv('Homicidios_antes_del_regimen.csv')

# Modificar los encabezados de las columnas
df.columns = ['AÑOS', 'TOTAL', 'MUJERES', 'HOMBRES', 'MUJERES_X_FAMILIA', 'HOMBRES_X_FAMILIA', 'HOMICIDIOS_POR_CADA_100_MIL_PERSONAS']

# Eliminar columnas innecesarias
df.drop(columns=['MUJERES_X_FAMILIA', 'HOMBRES_X_FAMILIA'], inplace=True)

# Eliminar filas con valores faltantes en columnas clave
df.dropna(subset=['TOTAL', 'HOMBRES', 'HOMICIDIOS_POR_CADA_100_MIL_PERSONAS'], inplace=True)

# Limpiar y convertir la columna 'TOTAL' a enteros
df['TOTAL'] = df['TOTAL'].astype(str).str.replace('.', '', regex=False).astype(int)

# Corregir valores erróneos en 'TOTAL'
df.loc[df['TOTAL'] == 4960, 'TOTAL'] = 496

# Limpiar y convertir la columna 'HOMBRES' a enteros
df['HOMBRES'] = df['HOMBRES'].astype(str).str.replace('.', '', regex=False).astype(int)

# Corregir valores erróneos en 'HOMBRES' (sin inplace=True)
df['HOMBRES'] = df['HOMBRES'].replace({4250: 425, 9440: 944})

# Convertir la columna 'MUJERES' a enteros
df['MUJERES'] = df['MUJERES'].astype(float).astype(int)

# Formatear la columna 'HOMICIDIOS_POR_CADA_100_MIL_PERSONAS' para mostrar dos decimales con coma
df['HOMICIDIOS_POR_CADA_100_MIL_PERSONAS'] = (
    pd.to_numeric(df['HOMICIDIOS_POR_CADA_100_MIL_PERSONAS'], errors='coerce')
    .map(lambda x: f"{x / 100:,.2f}".replace('.', ',') if pd.notna(x) else x)
)

# Mostrar los primeros 11 registros
df.head(11)
```

### 3. Visualización de tendencias
Visualizamos las tendencias de la expectativa de vida y los homicidios antes y después del régimen.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Datos de esperanza de vida
esperanza_vida_data = {
    "Fecha": [2022, 2021, 2020, 2019, 2018, 2017, 2016, 2015, 2014, 2013, 2012, 2011, 2010],
    "Esperanza de vida - Mujeres": [758, 751, 754, 767, 774, 774, 770, 765, 762, 761, 761, 761, 762],
    "Esperanza de vida - Hombres": [166, 566, 166, 868, 967, 966, 866, 766, 766, 667, 667, 967, 167],
    "Esperanza de vida": [847, 87, 437, 17, 357, 917, 767, 837, 997, 157, 267, 297, 257]
}

esperanza_df = pd.DataFrame(esperanza_vida_data)

# Datos de homicidios (históricos antes del régimen)
homicidios_antes_data = {
    "Fecha": [2021, 2020, 2019, 2018, 2017, 2016, 2015, 2014, 2013, 2012, 2011, 2010],
    "Número de Homicidios": [1085, 1341, 2398, 3346, 3962, 5276, 6656, 3921, 2513, 2594, 4371, 3987]
}

homicidios_antes_df = pd.DataFrame(homicidios_antes_data)

# Datos de homicidios recientes (2024)
homicidios_recientes = {
    "Mes": ["ENE.", "FEB.", "MAR.", "ABR.", "MAY.", "JUN.", "JUL."],
    "Delincuencia General": [1, 1, 4, 0, 6, 3, 4],
    "Intolerancia Social": [6, 5, 2, 3, 3, 3, 5],
    "Intolerancia Familiar": [2, 2, 7, 4, 7, 2, 6]
}

recientes_df = pd.DataFrame(homicidios_recientes)

# Gráfico 1: Evolución de la esperanza de vida
plt.figure(figsize=(10, 6))
for column in ["Esperanza de vida - Mujeres", "Esperanza de vida - Hombres", "Esperanza de vida"]:
    plt.plot(
        esperanza_df["Fecha"], 
        esperanza_df[column], 
        label=column, 
        marker="o"
    )
plt.title("Evolución de la Esperanza de Vida", fontsize=14)
plt.xlabel("Año")
plt.ylabel("Esperanza de Vida (años)")
plt.legend()
plt.grid(True)
plt.show()

# Gráfico 2: Número de homicidios anuales antes del régimen
plt.figure(figsize=(10, 6))
plt.bar(
    homicidios_antes_df["Fecha"], 
    homicidios_antes_df["Número de Homicidios"], 
    color="skyblue", 
    label="Homicidios Anuales"
)
plt.title("Número de Homicidios Anuales Antes del Régimen", fontsize=14)
plt.xlabel("Año")
plt.ylabel("Número de Homicidios")
plt.grid(True)
plt.show()

# Gráfico 3: Distribución de homicidios recientes (2024)
plt.figure(figsize=(10, 6))
width = 0.25
x = np.arange(len(recientes_df["Mes"]))
plt.bar(x - width, recientes_df["Delincuencia General"], width, label="Delincuencia General", color="red")
plt.bar(x, recientes_df["Intolerancia Social"], width, label="Intolerancia Social", color="orange")
plt.bar(x + width, recientes_df["Intolerancia Familiar"], width, label="Intolerancia Familiar", color="green")
plt.title("Distribución de Homicidios Recientes (2024)", fontsize=14)
plt.xlabel("Mes")
plt.ylabel("Cantidad")
plt.xticks(x, recientes_df["Mes"])
plt.legend()
plt.grid(True)
plt.show()
```
